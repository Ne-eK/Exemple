## Задание 1.
Загружаем таблицу

```python
import pandas as pd

df = pd.read_csv ('/content/kc-house-data.csv')
df.head()
```

Подключаем matplotlib

строим график распределения цен

```python
import matplotlib.pyplot as plt

plt.hist(df['price'], edgecolor = 'black')
plt.title('Распределение цен')
plt.xlabel('Цены')
plt.ylabel('Количество')
```

строим график распределения жилой площади

```python
plt.hist(df['sqft_living'], edgecolor = 'black')
plt.title('Распределение жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Количество')
```

строим график распределения года постройки

```python
plt.hist(df['yr_built'], edgecolor = 'black')
plt.title('Распределение года постройки')
plt.xlabel('Год строительства')
plt.ylabel('Количество')
```

основне выводы:
1. большиство квартир стоят до 1 млн.
2. наиболее часто встречающаяся площадь построек 2000 - 3000
3. самы старые постройки страше 100 лет
4. наибольшее количество построек моложе 2000 года

## Заданиие 2

распределение по наличию вида на набережную

```python
data = df['waterfront'].value_counts()
plt.pie(data.values,  autopct='%.1f%%')
plt.legend(['Нет береговой линии', 'Есть береговая линия']);
```

Менее 1% построек имеют вид на набережную

Строим график распределениея по этажам

```python
plt.hist(df['floors'], edgecolor = 'black')
plt.title('Распределение этажей домов')
plt.xlabel('Этажность')
plt.ylabel('Количество');
```

Строим график распределения по классу построек

```python
data = df['grade'].value_counts()
names = data.index
value = data.values
plt.bar(names, value)
plt.xticks(rotation=90);
```

Большинство построек имеют 7 класс

## Задание 3


исследование цен

```python
import seaborn as sns
sns.boxplot(x=df['price'], y=df['floors'], showfliers=False)
plt.xlabel('Цены')
plt.ylabel('Этажность')
plt.title('Взаимосвязь этажности и цены');
```

```python
sns.boxplot(x=df['price'], y=df['waterfront'], showfliers=False)
plt.xlabel('Цены')
plt.ylabel('Береговая линия')
plt.title('Взаимосвязь вида на набрежную и цены');
```

при наличии вида на набережную цены выше

```python
sns.boxplot(x=df['price'], y=df['grade'], showfliers=False)
plt.xlabel('Цены')
plt.ylabel('Класс')
plt.title('Взаимосвязь класса и цены');
```

чем выш класс дома тем выше цена
